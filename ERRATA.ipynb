No **Exemplo 4.2** a figura correta é mostrada abaixo:


```python
import matplotlib.pyplot as plt
import numpy as np

f1 = lambda x: x**2-2
f2 = lambda x: -np.log(x)

x = np.linspace(0.01, 2, 100)
plt.plot(x, f1(x), x, f2(x))

plt.title('Localização de raízes por gráficos')
plt.legend(['$f_1(x)=x^2-2$','$f_2(x)=-ln(x)$'])
plt.ylabel('Eixo $y$')
plt.xlabel('Eixo $x$')
plt.grid()
```
