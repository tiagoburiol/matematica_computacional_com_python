**Exemplo 4.2:** 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f1 = lambda x: x**2-2
f2 = lambda x: -np.log(x)

x = np.linspace(0.01, 2, 100)
plt.plot(x, f1(x), x, f2(x))

plt.title('Localização de raízes por gráficos')
plt.legend(['$f_1(x)=x^2-2$','$f_2(x)=-ln(x)$'])
plt.ylabel('Eixo $y$')
plt.xlabel('Eixo $x$')
plt.grid()
```

```python
def bissec(f, a, b, eps=0.001):
    err = b-a
    x_ant = a
    while err > eps:
        x = (a+b)/2.0
        if f(x) == 0:
            break
        elif f(x)*f(a) < 0:
            b = x
        else:
            a = x
        err = abs(x - x_ant)/abs(x)
        x_ant = x
        print ("x = %.4f"%x,
               "E = %.4f"%err) 

    return (x)

f = lambda x: x**2+np.log(x)-2
bissec(f,1.25,1.5,0.001)
```

**Exemplo 4.9:** 

```python
xi = np.linspace(0.05, 5)
plt.plot(xi, np.log(xi), xi, xi-2) 
plt.grid()
plt.show()
```

```python
def pontofixo(phi,x,eps):
    err = 10.0
    x_ant = x
    
    while err>eps:
        x = phi(x)
        err = abs(x-x_ant)/abs(x)
        x_ant = x
        print ("| x=%.9f"%x, "| err=%.9f"%err)
    return (x)

phi1 =  lambda x: np.log(x)+2
x = pontofixo(phi1, 3.0,0.0001)
print(x)
phi2 = lambda x: np.exp(x-2)
x = pontofixo(phi2, 0.5, 0.0001)
print(x)
```

```python
f = lambda x: np.log(x)+x-4.0
df = lambda x: 1./x +1  
x = 2
x_ant = x
err = 10.0

while err>0.0001:
    x = x - f(x)/df(x)
    err = abs(x-x_ant)/abs(x)
    x_ant = x
    print ("| x=%.6f"%x, "| err=%.6f"%err)
```

```python
err = 10.0
x0 = 1.0
x1 = 2.0

f = lambda x: np.log(x)+x-4.0

while err> 0.0001:
    x = (x0*f(x1)-x1*f(x0))/(f(x1)-f(x0))
    err = abs(x-x1)/abs(x) 
    x0 = x1
    x1 = x
    print ("x=%.9f"%x, "E=%.9f"%err)
```
