## 2. Funções e gráficos

**Exemplo 2.1:** Obtendo a imagem e a representação gráfica da função $P(n)=350+50n$ com $D_f=\{2,3,4,5,6,7\}$. 

```python
# Criamos a função
def P(n):
    return 350 + n*50
```

```python
# Calculamos os valores
Df = [2,3,4,5,6,7] # domínio
Imf = []           # lista vazia para a imagem

# para cada n calcula P(n) e adiciona 
for n in Df:
    Imf.append(P(n))
    print('P(',n,')=', Imf[-1])
```

```python
import matplotlib.pyplot as plt
```

```python
# Plotamos os pontos
plt.plot(Df, Imf, 'o')
plt.grid()
```

**Exemplo 2:** Plotando o gráfico de $f(x) = x^2-7x+3$ com 21 pontos entre 10 e 10.

```python
import numpy as np
```

```python
xi = np.linspace(-10,10, 21)
f = lambda x: x**2-7*x+3
plt.plot(xi,f(xi), '.-')
plt.grid()
```

```python
print('xi:', xi[0:5])
print('yi:', f(xi[0:5]))
```

**Exemplo 2.3:** Plotando o gráfico de $f(d) = 1.047d^3-31.415d^2+2672.369$.

```python
di = np.linspace(-20,40)
f = lambda d: 1.047*d**3-31.415*d**2+2672.369

plt.plot(di,f(di))
plt.grid()
```

**Exemplo 2.4:** Plotando mais algumas funções.

```python
xi = np.linspace(-10,10)
f0 = lambda x: np.exp(x)*np.sin(x)
f1 = lambda x: np.cos(x)*np.log(x+10.1)
f2 = lambda x: np.arctan(x)*np.sqrt(x+10)


fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(12, 4))


axs[0].plot(xi,f0(xi))
axs[0].grid()
         
axs[1].plot(xi,f1(xi))
axs[1].grid()

axs[2].plot(xi,f2(xi))
axs[2].grid()
```
